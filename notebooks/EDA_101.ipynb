### Missing Value Analysis

```python
missing_df = df.isnull().sum().reset_index()
missing_df.columns = ['column_name', 'missing_count']

missing_df['missing_percentage'] = (missing_df['missing_count'] / len(df)) * 100
missing_df.sort_values(by='missing_percentage', ascending=False, inplace=True)
```

```python
missing_df.head(20)
```

```python
print(f"Number of columns with no missing values: {len(missing_df[missing_df['missing_count'] == 0])}")
print(f"Number of columns with with less than 10% missing values: {len(missing_df[( missing_df['missing_percentage'] > 0) & (missing_df['missing_percentage'] <= 10)])}")
print(f"Number of columns with with 10% to 25% missing values: {len(missing_df[(missing_df['missing_percentage'] > 10) & (missing_df['missing_percentage'] <= 25)])}")
print(f"Number of columns with with 25% to 50% missing values: {len(missing_df[(missing_df['missing_percentage'] > 25) & (missing_df['missing_percentage'] <= 50)])}")
print(f"Number of columns with with more than 50% missing values: {len(missing_df[missing_df['missing_percentage'] > 50])}")
```

### Seperate Numerical and Categorical Columns

```python
cat_columns = df.select_dtypes(include=['object']).columns.tolist()
num_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print(f"Number of categorical columns: {len(cat_columns)}")
print(f"Number of numerical columns: {len(num_columns)}")

print("\nCategorical columns:")
for col in cat_columns:
    print(f" - {col}")

# print("\nNumerical columns:")
# for col in num_columns:
#     print(f" - {col}")
```

### Target Distribution Plot

```python
import matplotlib.pyplot as plt

df["TARGET"].value_counts().plot(kind="bar")

plt.title("Target Variable Distribution")
plt.xlabel("TARGET")
plt.ylabel("Count")
plt.show()
```

Each row represents a loan/application.

The model should learn from customer/application attributes and predict whether the customer may default.

Since TARGET = 1 is risky/default customer, our model should especially focus on identifying TARGET = 1 correctly.

So, only accuracy is not enough.

We will give more importance to:
- ROC-AUC
- Recall for TARGET = 1
- Precision
- F1-score
- KS statistic
